```python
import os
import subprocess
```

```python
input_directory = 'ligand_3d_structures'
output_directory = 'output_pdbqt_files'
```

```python
os.makedirs(output_directory, exist_ok=True)
```

```python
# Loop through all SDF files in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith(".sdf"):
        input_file_path = os.path.join(input_directory, filename)
        output_file_path = os.path.join(output_directory, filename.replace('.sdf', '.pdbqt'))

        # Construct and run the Open Babel conversion command
        command = f"obabel -isdf {input_file_path} -opdbqt -O {output_file_path} -h"
        subprocess.run(command, shell=True)
```

```python
files = os.listdir(output_directory)

# Filter and count files that end with .pdbqt
pdbqt_files_count = sum(1 for file in files if file.endswith('.pdbqt'))

print(f"There are {pdbqt_files_count} PDBQT files in the directory '{output_directory}'.")
```
